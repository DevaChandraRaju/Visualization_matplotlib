```python
# 5) ## Stack Plot ##########

days = [1,2,3,4,5]
sleeping = [7,8,8,9,6]
eating = [2,3,4,2,2]
working = [7,8,7,3,3]
playing = [8,5,5,10,13]

plt.plot([],[],color='m',label='Sleeping',linewidth=3)
plt.plot([],[],color='c',label='Eating',linewidth=3)
plt.plot([],[],color='r',label='Working',linewidth=3)
plt.plot([],[],color='k',label='Playing',linewidth=3)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel('Days')
plt.ylabel('Activity')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

```python
# 6) ### Pie charts ##################
days = [1,2,3,4,5]
sleeping = [7,8,8,9,6]
eating = [2,3,4,2,2]
working = [7,8,7,3,3]
playing = [8,5,5,10,13]

slices = [7,2,2,13]
activities = ['Sleeping','Eating','Working','Playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=60,
        shadow=True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('Stack Plot')
plt.show()
```

```python
## Fill between plots
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

index = pd.date_range(start='2000-1-1', end='2015-1-1', freq='M') ## freq = B --> for business days
N = len(index)
poisson = (stats.poisson.rvs(1000, size=(N,3))/100.0)
poisson.sort(axis=1)
df = pd.DataFrame(poisson, columns=['lwr', 'Rt', 'upr'], index=index)
print(df.columns)

plt.fill_between(df.index, df.lwr, df.upr, facecolor='blue', alpha=.2)
plt.plot(df.index, df.Rt, '.')
plt.show()
```

```python
#  a) spines /boarder color
#  b) tick labels color change
#  c) drawing a think center or reference point line

import matplotlib.pyplot as plt
import numpy as np
from datetime import date as dt
import pandas as pd
import requests # this is to request URL data
import matplotlib.dates as mdates # for date manipulation

def convert(x):
    try:
        return float(x)
    except:
        return x

def ch(x):
    #19-Sep-16
    m = {'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,\
    'Oct':10,'Nov':11,'Dec':12}
    y = x.split('-')
    return(date(int('20'+y[2]), m[y[1]], int(y[0])))
    
def graph_data(stock):
    
    fig = plt.figure()
    ax1 = plt.subplot2grid((1,1),(0,0))
                          
    url = 'https://www.google.com/finance/historical?output=csv&q='+stock
    resp = requests.get(url)
    data = resp.text.split('\n')
    fields = data[0].split(',')
    df_data = df_list = []
    
    for each in data[1:]:
        df_data = [convert(x) for x in each.split(',')]
        if len(df_data)==6:
            df_list.append(df_data)
            
    index = pd.date_range(start='2016-9-6', end='2017-8-22', freq='B')
    N=len(index)

    df_stocks = pd.DataFrame(df_list,columns=fields,index=index)
    
    ### This is to plot line for date fields[matplotlib inbuild function helps to draw a plot for date data columns]
    ax1.plot_date(df_stocks.index,df_stocks['Open'],'-',label='Price')
    ax1.plot([],[],lw=3,label='Gain',color='g',alpha=0.5)
    ax1.plot([],[],lw=3,label='Loss',color='r',alpha=0.5)
    ax1.axhline(33,color='black')
    
    #### Fill the color in between based on stated condition
    ax1.fill_between(df_stocks.index,df_stocks.Open,33,where=(df_stocks.Open > 33), facecolor='g',alpha=0.5)
    ax1.fill_between(df_stocks.index,df_stocks.Open,33,where=(df_stocks.Open < 33), facecolor='r',alpha=0.5)
    
    ##### Below logic rotates labels of x/y aixs to fit more date
    for label in ax1.get_xticklabels():
        label.set_rotation(45) 
    ax1.grid(True)
    ax1.spines['left'].set_color('b')
    ax1.spines['left'].set_lw(5)
    ax1.tick_params(axis='x',colors='b')
          
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    #plt.subplots_adjust(left=0.09,bottom=0.18,right=0.94,top=0.93,wspace=0.2,hspace=0)
    plt.show()
    
#graph_data('TSLA')
graph_data('EBAY')
```
